```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df
```

```python
plt.figure(figsize=(12,10))
plt.suptitle('first subplot')
```

```python
plt.subplot(121)
plt.title('first plot 121')
plt.hist(data=df,x='fare')
plt.subplot(122)
plt.title('first plot 122')
plt.boxplot(data=df,x='fare')
plt.tight_layout()
```

```python
plt.subplot(211)
plt.title('first plot 211')
plt.pie(data=df,x='fare')
plt.subplot(212)
plt.title('first plot 212')
plt.boxplot(data=df,x='fare')
plt.tight_layout()
```
